```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
label_ML = ['Clean Data', 'Noisy Data']
```

```python
accuracy_ML_ex1 = [0.7684, 0.7004] # KB
accuracy_ML_ex2 = [0.8452, 0.6670] # KE
accuracy_ML_ex3 = [0.8776, 0.7924] # XGB1
accuracy_ML_ex4 = [0.8020, 0.7389] # XGB2
accuracy_ML_ex5 = [0.8812, 0.8361] # svd XGB
```

```python
x_ML = np.arange(len(label_ML))
```

```python
width = 0.35

fig1, ax1 = plt.subplots()
plt.rcParams["figure.figsize"] = (10, 7)

rects1 = ax1.bar(x_ML - 2*width/3, accuracy_ML_ex1, width/4, label='KB')
rects2 = ax1.bar(x_ML - width/3, accuracy_ML_ex2, width/4, label='KE')
rects3 = ax1.bar(x_ML, accuracy_ML_ex3, width/4, label='XGB1')
rects4 = ax1.bar(x_ML + width/3, accuracy_ML_ex4, width/4, label='XGB2')
rects5 = ax1.bar(x_ML + 2*width/3, accuracy_ML_ex5, width/4, label='ensemble')

ax1.set_ylabel('Score')
ax1.set_title('Graph : Comparison of Score with Different Models', pad=20)
ax1.set_xticks(x_ML)
ax1.set_xticklabels(label_ML)
ax1.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax1.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)

fig1.tight_layout()
plt.show()
```
