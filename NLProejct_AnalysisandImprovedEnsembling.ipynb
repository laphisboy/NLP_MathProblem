```python
class Simple(nn.Module):
    # can decide on the hidden layer size by calling class
    def __init__(self, n_input, n_hidden, n_out):
        super(Simple, self).__init__()
        self.hidden_layer1 = nn.Linear(n_input, n_hidden) # ax + b
        self.hidden_layer2 = nn.Linear(n_hidden, int(n_hidden/2))
        self.hidden_layer3 = nn.Linear(int(n_hidden/2), int(n_hidden/4))
        self.out = nn.Linear(int(n_hidden/4), n_out)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.hidden_layer1(x)
        x = self.relu(x)
        
        x = self.hidden_layer2(x)
        x = self.relu(x)
        
        x = self.hidden_layer3(x)
        x = self.relu(x)
        
        x = self.out(x)
        return x
```

```python
model = Simple(888, 128, 37).to(DEVICE) # 2-16-1 model to start with

nn.init.xavier_uniform_(model.hidden_layer1.weight)
nn.init.xavier_uniform_(model.hidden_layer2.weight)
nn.init.xavier_uniform_(model.hidden_layer3.weight)
nn.init.xavier_uniform_(model.out.weight)
```
